**Import Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier  # Import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, average_precision_score, confusion_matrix, roc_auc_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Credit_card_fraud_analysis/creditcard.csv')
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions - 492


```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating two DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Train and Test Data Split**

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Creation - Logistic Regression**

```python
logistic_model = LogisticRegression(max_iter=1000)  # Increase max_iter
logistic_model.fit(X_train, Y_train)
```

**Model Evaluation - Logistic Regression**

```python
# Predictions on the test data
X_test_prediction_logistic = logistic_model.predict(X_test)

# Calculate Precision
precision_logistic = precision_score(Y_test, X_test_prediction_logistic)

# Calculate Recall
recall_logistic = recall_score(Y_test, X_test_prediction_logistic)

# Calculate F1-Score
f1_score_logistic = f1_score(Y_test, X_test_prediction_logistic)

# Calculate AUPRC (Area Under the Precision-Recall Curve)
auprc_logistic = average_precision_score(Y_test, X_test_prediction_logistic)

# Calculate Confusion Matrix
confusion_matrix_logistic = confusion_matrix(Y_test, X_test_prediction_logistic)

# Calculate ROC-AUC
roc_auc_logistic = roc_auc_score(Y_test, X_test_prediction_logistic)

# Print the metrics
print(f"Logistic Regression Metrics:\nPrecision: {precision_logistic:.4f}\nRecall: {recall_logistic:.4f}\nF1-Score: {f1_score_logistic:.4f}\nAUPRC: {auprc_logistic:.4f}\nConfusion Matrix:\n{confusion_matrix_logistic}\nROC-AUC: {roc_auc_logistic:.4f}")
```

**Model Creation - Random Forest**

```python
random_forest_model = RandomForestClassifier(n_estimators=100)  # Initialize RandomForestClassifier
random_forest_model.fit(X_train, Y_train)
```

**Model Evaluation - Random Forest**

```python
# Predictions on the test data
X_test_prediction_rf = random_forest_model.predict(X_test)

# Calculate Precision
precision_rf = precision_score(Y_test, X_test_prediction_rf)

# Calculate Recall
recall_rf = recall_score(Y_test, X_test_prediction_rf)

# Calculate F1-Score
f1_score_rf = f1_score(Y_test, X_test_prediction_rf)

# Calculate AUPRC (Area Under the Precision-Recall Curve)
auprc_rf = average_precision_score(Y_test, X_test_prediction_rf)

# Calculate Confusion Matrix
confusion_matrix_rf = confusion_matrix(Y_test, X_test_prediction_rf)

# Calculate ROC-AUC
roc_auc_rf = roc_auc_score(Y_test, X_test_prediction_rf)

# Print the metrics
print(f"Random Forest Metrics:\nPrecision: {precision_rf:.4f}\nRecall: {recall_rf:.4f}\nF1-Score: {f1_score_rf:.4f}\nAUPRC: {auprc_rf:.4f}\nConfusion Matrix:\n{confusion_matrix_rf}\nROC-AUC: {roc_auc_rf:.4f}")
```
